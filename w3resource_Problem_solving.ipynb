```python
def odd_even(numbers):
  odd_numbers=[]
  even_numbers=[]
  for num in numbers:
    if num % 2==0:
      odd_numbers.append(num)
    else :
      even_numbers.append(num)
  return odd_numbers,even_numbers
```

```python
def odd_even2(numbers):
    odd_list = []
    even_list = []
    for num in numbers:
        if num % 2 == 0:
            even_list.append(num)
        else:
            odd_list.append(num)
    return odd_list, even_list
```
